```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configure visualization aesthetics
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({'figure.figsize': (10, 6), 'axes.titlesize': 14})

print("EDA Environment Initialized. Visualization libraries configured.")
```

```python
PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd()
PROCESSED_DATA_PATH = PROJECT_ROOT / 'data' / 'processed' / 'RTA_cleaned.csv'

try:
    df = pd.read_csv(PROCESSED_DATA_PATH)
    print(f"Successfully loaded cleaned dataset. Shape: {df.shape}")
except FileNotFoundError:
    print("CRITICAL ERROR: Cleaned dataset not found. Ensure 02_cleaning.ipynb was executed.")
    raise
```

```python
ax = sns.countplot(data=df, x='Accident_severity', order=df['Accident_severity'].value_counts().index)
plt.title("Baseline Distribution of Accident Severity")
plt.ylabel("Number of Accidents")
plt.xlabel("Severity Level")

# Add value labels to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

print("Generating target variable distribution...")
plt.show()
```

```python
print("Generating Accident Severity across Patrol Shifts...")
# Accident Severity across Patrol Shifts
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Patrol_Shift', hue='Accident_severity')
plt.title("Accident Severity by Patrol Shift")
plt.ylabel("Accident Count")
plt.xlabel("Operational Shift")
plt.xticks(rotation=15)
plt.legend(title='Severity')
plt.show()

print("Generating Impact of Adverse Environmental Conditions...")
# Impact of Adverse Environmental Conditions
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Is_Adverse_Condition', hue='Accident_severity')
plt.title("Impact of Adverse Weather/Road Conditions on Accident Severity")
plt.ylabel("Accident Count")
plt.xlabel("Adverse Conditions Present?")
plt.legend(title='Severity')
plt.show()
```

```python
print("Generating Demographic Risk Analysis...")

# Driving Experience vs. Accident Severity
plt.figure(figsize=(10, 6))
# Filter out 'Unknown' to keep the chart clean
mask = df['Driving_experience'] != 'Unknown'
sns.countplot(data=df[mask], y='Driving_experience', hue='Accident_severity', 
              order=df[mask]['Driving_experience'].value_counts().index)
plt.title("Accident Severity by Driver's Experience Level")
plt.xlabel("Accident Count")
plt.ylabel("Driving Experience")
plt.legend(title='Severity', loc='lower right')
plt.show()
```

```python
print("Generating Infrastructure Risk Analysis...")

# Junction Type vs. Severity
plt.figure(figsize=(12, 6))
mask = df['Types_of_Junction'] != 'Unknown'
sns.countplot(data=df[mask], y='Types_of_Junction', hue='Accident_severity',
              order=df[mask]['Types_of_Junction'].value_counts().index)
plt.title("Accident Severity by Junction Type")
plt.xlabel("Accident Count")
plt.ylabel("Junction Type")
plt.legend(title='Severity', loc='lower right')
plt.show()
```
